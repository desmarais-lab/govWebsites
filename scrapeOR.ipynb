```python
from random import gauss
from time import sleep
cand_IDs = []
path = '/home/markus/govWebsites/data/OR_mayors.txt'
with open(path) as f:
    for line in f:
        url = "https://secure.sos.state.or.us/orestar/GotoSearchByName.do"
        browser.get(url)
        sleep(gauss(7, .5))
        elem = browser.find_element_by_name("committeeName")
        elem.clear()
        elem.send_keys(str(line.strip()))
        elem = browser.find_element_by_name("submit")
        elem.click()
        sleep(gauss(7, .5))
        try:
            cand_id = browser.find_element_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "evenRowTop", " " )) and (((count(preceding-sibling::*) + 1) = 2) and parent::*)]')
            cand_IDs.append(cand_id.text)
        except Exception as e:
            print(e)
            cand_IDs.append(999999999)
            continue
```
